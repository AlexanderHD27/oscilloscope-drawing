```python
FP_SHIFT = 16
FP_MOD_MASK = 2**32

FP_FROM_UINT = lambda x: (x << FP_SHIFT) % FP_MOD_MASK
FP_TO_INT16 = lambda x: (x >> FP_SHIFT) % 2**16

FP_MUL = lambda x, y: (((x * y) % 2**64) >> FP_SHIFT) % FP_MOD_MASK
FP_DIV = lambda x, y: int((x << FP_SHIFT) / y) % FP_MOD_MASK 

def print_fp32(x: int):
    print("{:08x}".format(x % FP_MOD_MASK))
```

```python
total = 1000
for i in range(total):
    print_fp32(FP_MUL(FP_DIV(FP_FROM_UINT(i), FP_FROM_UINT(total)), FP_FROM_UINT(0xffff)))
```

```python
print_fp32(FP_MUL(FP_FROM_UINT(2), FP_FROM_UINT(0x7fff)))
```
